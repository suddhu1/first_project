```python
cut_categories=['Premium', 'Very Good', 'Ideal', 'Good', 'Fair']
color_categories=['F', 'J', 'G', 'E', 'D', 'H', 'I']
clarity_categories=['VS2', 'SI2', 'VS1', 'SI1', 'IF', 'VVS2', 'VVS1', 'I1']
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline =Pipeline(
    steps=[
        ('imputer',SimpleImputer(strategy='median')),
        ('scaler',StandardScaler()),

        

    ]
)
cat_pipeline=Pipeline(
    steps=[
        ('imputer',SimpleImputer(strategy="most_frequent")),
        ('encoding',OrdinalEncoder(categories=[cut_categories,color_categories,clarity_categories])),
    ]
)

preprocessor =ColumnTransformer(
    [
        ('num_pipeline',num_pipeline,numerical_column),
        ('categorical_pipeline',cat_pipeline,categorical_column)
    ]
)
```

```python
import numpy as np 
def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2_square = r2_score(y_true, y_pred)
    return mae, rmse, r2_square
```

```python
## Train multiple models 

models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'Elasticnet':ElasticNet(),
    'SVR':SVR(),
   'DTR': DecisionTreeClassifier()
}
trained_model_list=[]
model_list=[]
r2_list=[]

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(x_train,y_train)

    # make predictions 
    y_pred=model.predict(x_test)

    mae, rmse, r2_square=evaluate_model(y_test,y_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])

    print('Model Training Performance ')
    print('RMSE:',rmse)
    print('MAE:',mae)
    print('MAE:',mae)
    print("R2 score",r2_square*100)

    r2_list.append(r2_square)

    print('='*35)
    print('\n')
```
